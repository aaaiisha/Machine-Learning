```python
#DATAFRAME
df = pd.DataFrame(xarr,columns=['d1','d2'])
df
```

```python
#standardize 
dfstd  = (df - df.mean()) / (df.std())
print(dfstd)
cov_mat = np.cov(dfstd.T, bias = 0)
print('MATRIX:',cov_mat)
print('VARIANCE ',((dfstd.d1)**2).sum()/4)
print('COVARIANCE',((dfstd.d1)*(dfstd.d2)).sum()/4)
eigen_val, eigen_vect = np.linalg.eig(cov_mat)
print('EIGEN VALUES:',eigen_val)
print('EIGEN VECTORS',eigen_vect)
```

```python
top=eigen_vect[:,:2]
print(top)
td = np.matmul(np.array(dfstd),top)
print(td)
pcsdf=pd.DataFrame(data = td,columns = ['PC1','PC2'])
pcsdf
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(df)
pComponents = pca.fit_transform(df)
skldf = pd.DataFrame(data = pComponents
             , columns = ['PC1','PC2'])
print(pca.components_)
print(pca.explained_variance_)
skldf
```

```python
def ortopro(v0, v1, ax=None):
    ax = ax or plt.gca()
    arrowprops=dict(arrowstyle='->',
                    linewidth=2,
                    shrinkA=0, shrinkB=0)
    ax.annotate('', v1, v0, arrowprops=arrowprops)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 6))
fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.1)

# plot orig
ax[0].scatter(xarr[:, 0], xarr[:, 1], alpha=0.9)
for length, vector in zip(pca.explained_variance_, pca.components_):
    v = vector * 3 * np.sqrt(length)
    ortopro(df.mean(axis=0), df.mean(axis=0) + v, ax=ax[0])

ax[0].set(xlabel='x', ylabel='y', title='original')

# plot PC's
df_pca = pca.transform(df)
ax[1].scatter(df_pca[:, 0], df_pca[:, 1], alpha=0.9)
ortopro([0, 0], [0, 3], ax=ax[1])
ortopro([0, 0], [3, 0], ax=ax[1])
ax[1].set(xlabel='PC1', ylabel='PC2',
          title='PCs',
          xlim=(-5, 5), ylim=(-3, 3.1))
```
