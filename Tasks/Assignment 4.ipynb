```python
from sklearn.datasets import load_breast_cancer
d = load_breast_cancer()
#print(d.keys())
#print(d.feature_names)
#print (len(d.feature_names))
print (len(d.data))
print ("Feature shape:",len(d.data),len(d.feature_names))
list(d.target_names)

```

```python
def targets():
    data = numpy.c_[d.data, d.target]
    columns = numpy.append(d.feature_names, ["target"])
    return pandas.DataFrame(data, columns=columns)

def maindata():
    datadf = targets()
    X = datadf[datadf.columns[:-1]]
    y = datadf.target
    return X, y



x, y = maindata()
assert x.shape == (569, 30)
assert y.shape == (569,)
targets()
```

# divide the dataset for train and test
train size 70% and test size 30%

```python
from sklearn.model_selection import train_test_split

def traintest():
    X, y = maindata()
    return train_test_split(X, y, test_size=0.30, random_state=0)

x_train, x_test, y_train, y_test = traintest()
traintest()
```

```python
#my function num 2
def perf_measure(y_actual, y_pred):
    TP = 0
    FP = 0
    TN = 0
    FN = 0
    for i in range(len(y_pred)): 
        if y_actual[i]==y_pred[i]==1:
           TP += 1
        if y_pred[i]==1 and y_actual[i]!=y_pred[i]:
           FP += 1
        if y_actual[i]==y_pred[i]==0:
           TN += 1
        if y_pred[i]==0 and y_actual[i]!=y_pred[i]:
           FN += 1
    #return(TP, FP, TN, FN)
    return ((TP+TN)/(TP+TN+FP+FN))*100

perf_measure(actual,pred)
```
